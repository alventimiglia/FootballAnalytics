```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from mplsoccer.pitch import VerticalPitch
```

```python
#set general use colors
text_color = 'w'
```

```python
data = pd.read_csv('/Users/annaventimiglia/Downloads/shotmaps.csv')
```

```python
data
```

```python
fig,ax = plt.subplots(figsize=(13,8.5))
fig.set_facecolor('#22312b')
ax.patch.set_facecolor('#22312b')

pitch = VerticalPitch(pitch_type='statsbomb', half=True,
              pitch_color='#22312b', line_color='#c7d5cc',
              axis=True, label=True)

pitch.draw(ax=ax, tight_layout=False, constrained_layout=True)

plt.gca().invert_yaxis()

plt.scatter(data['x'],data['y'],s=100)

plt.title('Barcelona Shots vs Juventus', fontsize=24, c=text_color)
```
