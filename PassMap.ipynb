```python
import pandas as pd
import matplotlib.pyplot as plt
from mplsoccer.pitch import Pitch
import seaborn as sns
```

```python
pitch = Pitch(pitch_color='grass', line_color='white', stripe=True)
fig, ax = pitch.draw()
```

```python
#Read the data
df = pd.read_csv('/Users/annaventimiglia/Downloads/messibetis.csv')
```

```python
df
```

```python
#Matching the coordinates appropriately to fit in the dimensions of the pitch (instead of 100x100, 120x80)
df['x'] = df['x']*1.2
df['y'] = df['y']*0.8
df['endX'] = df['endX']*1.2
df['endY'] = df['endY']*0.8
```

```python
df
```

```python
fig,ax = plt.subplots(figsize=(13.5,8))
fig.set_facecolor('#22312b')
ax.patch.set_facecolor('#22312b')

#from documentation for mplsoccer
pitch = Pitch(pitch_type='statsbomb', half=False,
              pitch_color='#22312b', line_color='#c7d5cc',
              axis=True, label=True)
pitch.draw(ax=ax, tight_layout=False, constrained_layout=True)
#Inverting y-axis due to how the statsbomb pitchtype is drawn - see documentation
plt.gca().invert_yaxis()

#Making a for loop to go through each row of the dataframe
for x in range(len(df['x'])):
    if df['outcome'][x] == 'Successful':
        plt.plot((df['x'][x],df['endX'][x]),(df['y'][x],df['endY'][x]),color='green')
        #Adding points for the start position of the pass
        plt.scatter(df['x'][x],df['y'][x],color='green')
    if df['outcome'][x] == 'Unsuccessful':
        plt.plot((df['x'][x],df['endX'][x]),(df['y'][x],df['endY'][x]),color='red')
        plt.scatter(df['x'][x],df['y'][x],color='red')

plt.title('Messi Pass Map vs. Real Betis',color='white',size=20)
```
